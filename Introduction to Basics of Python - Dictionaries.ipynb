## Dictionaries

A dictionary is a data type built into Python. It is a collection of key-value pairs, which is unordered, changeable and indexed. 

A dictionary has following properties:
* keys within a dictionary are unique
* keys are either number or strings
* keys are case sensitive
* keys are immutable

Python provides a wide variety of built-in functions and methods to interact with dictionaries.

OK! So we have one dictionary object, one list, what is the difference?

#### Traversing a dictionary

* 'for' or 'while' loop can be used to iterate on a dictionary
* iteration on keys can happen on either the dictionary object or view returned by keys() method
* iteration on values can happen on the view returned by values() method
* iteration on key-value pairs can happen on the view returned by items() method

Of course iteration can happen on the respective lists created using list() on any of these views.

```python
for k in d1:
    print(k)
```

```python
for k in d1.keys():
    print(k)
```

```python
for v in d1.values():
    print(v)
```

```python
for k,v in d1.items():
    print(k,v)
```

```python
for n in d1.items():
    print(n)
```

```python
for k in list(d1.keys()):
    print(k)
```

At least no difference found in iteration over the view and the list created on the view!!

## The 'del' Statement

Used to delete a key-value pair from a dictionary.

Syntax: del dictionary_name[key]

```python
d3 = {1.1:'Create', 1.2:'Access', 1.3:'Modify'}
d3
```

```python
del d3[1.2]

d3
```

Let's see the difference of dictionary object and true list on dictionary keys here.

```python
for k in d1.keys():
    if k == 2:
        del d1[k]
```

```python
for k in list(d1.keys()):
    if k == 2:
        del d1[k]

d1
```

dict.keys() returns just a view of the keys of the dictionary, so while we can iterate over this view, we can't modify the dictionary at the same time. 

list(dict.keys()) returns true list, over which we can iterate and then use the key value to modify the dictionary as well.

## Sample Python Code

```python
# Program to Demonstrate Nested Dictionaries

# Program will use user defined functions, for loop, if statements

student_details = [{"name": "jasmine", "registration_number": "MOD1", "sub_marks": {"python": 95, "java": 90, ".net": 85}},
                   {"name": "daisy", "registration_number": "MOD2", "sub_marks": {"python": 93, "java": 89, ".net": 75}}]

def nested_dictionary():
    for student in student_details:
        print(f"Student Name {student['name']}")
        print(f"Registration Number {student['registration_number']}")
        average = sum(student["sub_marks"].values()) / len(student["sub_marks"])
        print(f"Average of all the subjects is {average}")

def main():
    nested_dictionary()

if __name__ == "__main__":
    main()
```
