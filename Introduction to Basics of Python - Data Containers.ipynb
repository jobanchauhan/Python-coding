## List vs Tuples

- Tuples are immutable, and usually, contain a heterogeneous sequence of elements that are accessed via unpacking or indexing.

- If an item within a tuple is mutable, then you can change it. A list present as an item in a tuple, then any changes to the list get reflected on the overall items in the tuple.

## Tuples as Return Values

Tuples can also be returned by the function as return values. Generally, the function returns only one value but by returning tuple, a function can return more than one value.

```python
def number_string_length(string_list, k):
    string_length = []
    
    for s in string_list:
        string_length.append(len(s))
        
    return (k, string_length.count(k))
```

```python
number_string_length(['DS', 'ML', 'KD', 'MLDL', 'KNN'], 2)
```

## Frozenset

- A frozenset is basically the same as a set, except that it is immutable. 


- Once a frozenset is created, then its items cannot be changed. 


- Since they are immutable, they can be used as members in other sets and as dictionary keys. 


- The frozensets have the same functions as normal sets, except none of the functions that change the contents (update, remove, pop, etc.) are available.

## Example 1

Write a Python Program to search a given number in a list and print Success or Failure.

```python
number_items = int(input('Enter number of elements in list '))
number_list = []

for i in  range(1, number_items+1):
    num = int(input(f'Enter number #{i} '))
    number_list.append(num)

print('\n List elements are : ', number_list, sep=' ')

search_item = int(input('Enter number to search in the list. '))
found = False

for i in range(number_items):
    if(number_list[i] == search_item):
        print(f'\nNumber found at index {i+1}.')
        found = True
        break

if not found:
    print('\nNumber not in list.')
```

## Example 2

Write a Python Program to sort a string list by decreasing order of length.

```python
string_list = ['machine', 'learn', 'deeplearning', 'DS', 'mean', 'stddev', 'kurtosis', 'skewed']

string_size_list = []

for s in string_list:
    string_size_list.append((s, len(s)))
    
print(f'String list : \n {string_size_list}')

sorted_list = sorted(string_size_list, key=lambda s : s[1], reverse=True)

print(f'\n\n Sorted string list : \n{sorted_list}')
```

## Example 4

Mean partitioning. Given a list of numbers, partition list in two sets such that one list has all the numbers above mean and second below mean.

```python
num_list = [randint(1, 300) for j in range(50)]
```

```python
print(num_list)
```

```python
list_sum = 0

for s in num_list:
    list_sum += s
    
mean_list = list_sum/len(num_list)

print(mean_list)
```

```python
above_avg = []
below_avg = []

for e in num_list:
    if e >= mean_list:
        above_avg.append(e)
    else:
        below_avg.append(e)

print(above_avg)

print(below_avg)
```
